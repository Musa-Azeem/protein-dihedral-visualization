```python
from Bio.PDB import PDBParser
import numpy as np
```

```python
af_pdb = '/home/musa/dev/protein-dihedral-visualization/alphafold_predictions/6t1z.pdb'
parser = PDBParser()
af_structure = parser.get_structure('af', af_pdb)
chain = next(iter((af_structure[0].get_chains())))
residues = list(chain.get_residues())
for residue in residues:
    print(residue.resname, [atom for atom in residue])
    ave_bfactor = np.mean([atom.bfactor for atom in residue])
    print(ave_bfactor)
```
